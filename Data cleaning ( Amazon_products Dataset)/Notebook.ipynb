```python
'''     Tasks: 
        1. Deleting redundant columns
        2. Renaming the columns
        3. Dropping duplicates
        4. Remove the NaN values from the dataset
        5. Check for some more Transformations
'''
```

# 1. Deleting redundant columns

```python
data.columns
```

```python
columns_to_drop = ['Unnamed: 0', 'asin','product_url', 'product_photo','product_num_offers','product_minimum_offer_price','is_amazon_choice','is_prime', 'climate_pledge_friendly','sales_volume', 'delivery', 'has_variations', ]
```

```python
data.drop(columns = columns_to_drop,inplace = True)
```

```python
data.head(5)
```

# 2. Renaming the columns

```python
data.head(4)
```

```python
data_new_columns=[]
for col in data.columns:
    data_new_columns.append(col.capitalize())
```

```python
data.columns = data_new_columns
```

```python
data.columns
```

# 3. Dropping duplicates

```python
data.duplicated().value_counts()
```

```python
data.duplicated().sum()
```

```python
# there are no duplicates in the data
```

# 4. Remove the NaN values from the dataset

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

#  5. Check for some more Transformations

```python
data.head(5)
```

```python
data["Is_best_seller"] = data["Is_best_seller"].apply(lambda x: 1 if x == True else 0)
```

```python
data["Is_best_seller"]
```

```python
data["Product_price"] = data["Product_price"].str.replace("$","")
```

```python
data["Product_price"].astype(float)
```

```python
data["Product_price"] = data["Product_price"].astype(float)
```

```python
data
```

```python
type(data["Product_price"][0])
```

```python
data["Product_price"] = data['Product_price'].apply(lambda x: round(x))
```

```python
data["Product_price"]
```
