# 1. Deleting redundant columns

```python
data.columns
```

```python
columns_to_drop =['Unnamed: 0','TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild',]
```

```python
data.drop(columns = columns_to_drop ,inplace = True)
```

```python
data.columns
```

# 2. Renaming the columns

```python
data.columns
```

```python
# changing name 'NrSiblings' to 'NumberofSiblings'
```

```python
data.rename(columns = {'NrSiblings':'NumberofSiblings'},inplace=True)
```

```python
data
```

# 3. Dropping duplicates

```python
data.duplicated()
```

```python
data.duplicated().value_counts()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.duplicated().value_counts()
```

# 4. Remove the NaN values from the dataset

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

#  5. Check for some more Transformations

```python
data
```

```python
data['Gender'] = data['Gender'].apply(lambda x: 'F' if x == 'Female' else 'M')
```

```python
data['Gender']
```

```python
data.reset_index(inplace = True)
```

```python
data
```

```python
data.drop(columns = ['index'] ,inplace=True)
```

```python
data
```

```python
data['NumberofSiblings'] = data['NumberofSiblings'].astype('int')
```
