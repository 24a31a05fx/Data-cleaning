# 1. Deleting redundant columns

```python
data.columns
```

```python
columns_to_keep = ['Age', 'Gender','Marital Status', 'Occupation', 'Monthly Income','Educational Qualifications', 'Family size','Feedback', 'Unnamed: 12']
columns_to_drop = ['latitude', 'longitude','Pin code','Unnamed: 12']
```

```python
len(columns_to_keep)
```

```python
len(columns_to_drop)
```

```python
data.head()
```

## 1st method : Filtering the columns

```python
modified_data  = data[columns_to_keep]
```

```python
modified_data
```

## 2nd method : Dropping the unnecessary columns

```python
data.drop(columns = columns_to_drop , inplace = True)
```

```python
data
```

# 2. Renaming the columns

```python
data.columns
```

```python
data.rename(columns = {'Monthly Income' : 'Monthly Salary'})
```

# 3. Dropping duplicates

```python
data.duplicated().value_counts()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.duplicated().value_counts()
```

# 4 . Remove the NaN values from the dataset

```python
data.isna().sum()
```

```python
# No NaN values in the dataset
```

# 5. Check for some more Transformations

```python
data
```

```python
# changing the age ( years) to age ( months)
data['Age (Months)'] = data ['Age']*12
```

```python
data
```

```python
# changing the gender female to F and male to M
```

```python
data['Gender'] = data['Gender'].apply(lambda x:'F' if x == 'Female' else 'M')
```

```python
data['Gender']
```

```python
data.head()
```

```python
# If family size > 3 its going to be large else small
```

```python
data['Family size'] = data['Family size'].apply(lambda x: 'Large' if x>3 else 'Small')
```

```python
data['Family size']
```

```python
# changing output values 'yes' - 1 and 'no' - 0
```

```python
data['Output'] = data['Output'].apply(lambda x: 1 if x=='Yes' else 0 )
```

```python
data['Output']
```

# Exporting the cleaned data into csv and excel files

```python
data.to_csv("Cleaned_data_csv.csv")
```
